1. How is the sampling frequency defined?

**Ans.** The rate at which a continuous signal is sampled to create a discrete signal is known as the sampling frequency, and it is commonly represented as f s. It shows how many samples are captured in a second and is given in Hertz (Hz). It is defined as
$$
 f_s = 1/T_s $$
 where:

*  f_s is the sampling freqency
*  T_s is the sampling period


  The Nyquist-Shannon sampling theorem states that the sampling frequency needs to be at least twice the greatest frequency component of the signal in order to reliably and alias-free reconstruct a signal from its samples:
$$
f_s \geq 2f_{\max} \tag{1}
$$


  where, fmax is the maximum freqency of the original signal



2. What happens to the output signal when under/oversampling?
**Ans.** When oversampling and undersampling occurs in the signal it is very hard to reconstruct the original signal.




*    Undersampling: When sampling frequency is less then twice the highest frequency of the signal then oversampling case occurs in the signal
$$ f_s <  2f_\max $$

  When undersampling happens in the signal at that time aliasing effect is shown up. During this issue high-frequency components of the signal are misrepresented as lower frequencies the signal became distorted and reconstruction to original signal is compromised.
*  Oversampling: When sampling frequency is greater than Nyquiest rate. $$ f_s >>  2f_\max $$

  It reduces the quantization noise, helps to improve Signal-to-Noise ratio, no need for analog anti-aliasing filkter which reduces the cost of filters. Provide better resolution and allow us to detect signal fast.

3. How do you determine the sampling frequency for a given signal? What happens during superposition of multiple signals?
**Ans.** Using the Nyquist-Shannon sampling theorem, you must take into account the signal's highest frequency component in order to calculate the sampling frequency.  According to this theory, in order to prevent aliasing, the sampling frequency must be at least twice the highest frequency in the signal.

The total of the various signals at each moment in time is the final signal produced when several signals are superimposed. Since the amplitudes of the waves can be directly summed, the superposition principle assumes that the waves are linear. Superposition can occasionally result in constructive or destructive interference, depending on whether the wave amplitudes reinforce or cancel one other out. A window function, sometimes referred to as an smoothening function or tapering function, is a mathematical function that has a zero value outside of a selected interval in signal processing and statistics. Window functions usually taper away from the interval's center, reach a maximum there, and are symmetric about the interval's center. When a window function is mathematically multiplied by another function, waveform, or data sequence, the product is also zero-valued outside the interval the view through the window is the only area that remains. In reality, this means that only the data segment inside the window is multiplied by the window function values after it has been isolated. Therefore, the primary role of window functions is tapering rather than segmentation.
