**Question 1**

 Define the Short Time Fourier Transform (STFT) and explain its purpose in signal processing. How does it differ from the regular Fourier Transform?


  **Ans:**


Defination: The Short Time Fourier Transform (STFT) is a mathematical tools used to analyze the content of frequency of non-stationary signals (A statinary signal is a signal whose statistical properties like mean, Frequency content, Amplitude changes over time being ) in digital processing. This method is used to represents the signal into time and frequency domain at a time.

The STFT for a signal called x(t) defined as

$$
\text{STFT}\{x(t)\}(t,\omega) = X(t,\omega) = \int_{-\infty}^{\infty} x(\tau) w(\tau - t) e^{-j \omega \tau} \, d\tau
$$


where:

**STFT Formula Components**


* x(t): the input signal  
* w(t): a window function (e.g., Hamming, Rectangle) centered at time \( t \)  
* 𝜔 (or 𝑓 for frequency) is the angular frequency.
* 𝑋(𝑡, 𝜔) is the STFT of the signal 𝑥(𝑡)

Applications:


STFT is widely used in various field, for example
* Speech processing: Analyzing and recognizing speech pattern.
* Audio Processing: Analyzing and manipulating music signals.
* Sonar: Analyzing underwater sounds.
* Mechanical Vibration analysis: To track and identify the harmonics components in noise signals.
* Radio communication: To Displaing and analyzing the radio signals.
* Fault detection in mechanical systems.

The key difference between an ordinary Fourier Transform and STFT is that in STFT method we only analyze our signsl in short overlapping time segments while in other method we thoroughly examine our signal in one go. The creation of a spectrogram relies heavily on the **Short-Time Fourier Transform (STFT)**. The STFT is a technique used to determine the frequency components of localized sections of a longer signal. The relationship is as follows:

1.  **Signal Segmentation (Windowing):** The STFT breaks down the main signal into smaller, often overlapping, segments. Each small segment is typically multiplied by a "window function" (like a Hann or Hamming window). This step helps to minimize spectral artifacts by smoothly tapering the edges of each segment before analysis .
2.  **Frequency Analysis (FFT):** The Fast Fourier Transform (FFT), an efficient algorithm to compute the Discrete Fourier Transform, is then applied to each of these windowed segments. This transforms each short time-slice of the signal from the time domain into the frequency domain, showing the amplitude and phase of the frequencies it contains.
3.  **Spectrogram Construction:** The STFT process results in a series of frequency spectra, one for each analyzed segment. The **spectrogram is formed by arranging these individual spectra side-by-side**. Usually, it plots the magnitude squared (which represents power) or just the magnitude of the frequency components found by the STFT. Each vertical line or column in the spectrogram image corresponds to the frequency spectrum of one short segment of the signal.<a name="sampling_anchor"></a>  
[<sup>[3]</sup>](#reference3)

**Visualizing Time-Varying Frequency Content:**

The spectrogram is powerful because it allows us to observe the evolution of a signal's frequency makeup:
* The **time axis** (usually horizontal) lets you track when changes occur.
* The **frequency axis** (usually vertical) shows what frequencies are present.
* The **color or intensity** at any specific time-frequency point indicates how prominent that frequency is at that moment.

For example, analyzing speech with a spectrogram can reveal how resonant frequencies, known as formants, shift as different vowels are spoken. In music, it can visually distinguish melodies, harmonies, and the characteristic sounds of various instruments over a piece's duration. The choice of the segment length (window size) in the STFT is important: short segments offer precise timing of frequency events but blur distinct frequencies, while longer segments give clearer frequency detail but less precise timing.[<sup>[3]</sup>](#reference3)

**Question 4**

What is the Phase Vocoder and what is its role in time stretching or pitch shifting audio signals? Explain the basic principles behind the Phase Vocoder algorithm.

**Ans:**

The **Phase Vocoder (PV)** is a sophisticated audio processing algorithm that operates in the frequency domain, primarily using the Short-Time Fourier Transform (STFT). It's designed to analyze, modify, and then resynthesize an audio signal, allowing for independent adjustments to its duration and pitch. The concept was pioneered in early digital signal processing.[<sup>[4]</sup>](#reference4) Its main applications are **time stretching** (making audio longer or shorter without changing its pitch) and **pitch shifting** (making the pitch higher or lower without changing its duration).

**Role in Time Stretching and Pitch Shifting:**

* **Time Stretching:** To change an audio signal's length without altering its pitch, the Phase Vocoder analyzes the input sound with a specific STFT window advancement (analysis hop size). It then reconstructs the sound using a *different* window advancement (synthesis hop size). If the synthesis hop is larger than the analysis hop, the sound becomes longer; if smaller, the sound becomes shorter. A critical part of this process is the careful adjustment of phase information from one analyzed segment to the next during reconstruction. This phase coherence is vital to prevent audible artifacts, an area where significant improvements have been made.[<sup>[3]</sup>](#reference3)
* **Pitch Shifting:** To change the pitch without altering duration, a common Phase Vocoder strategy involves a two-step process: first, time-stretch the signal by a factor corresponding to the desired pitch shift, and then resample the stretched signal (which inherently changes both pitch and duration) back to its original length. For instance, to double the perceived pitch, the sound might be time-stretched to twice its original length, and then played back at double speed.

**Basic Principles Behind the Phase Vocoder Algorithm:**

The Phase Vocoder algorithm typically works through three main stages:

1.  **Analysis:**
    * The incoming audio is divided into short, overlapping segments using a window function.
    * An FFT is performed on each segment. This yields the magnitude and phase for various frequency components within that short time window. Repeating this for all segments gives the STFT of the signal. [<sup>[3]</sup>](#reference3)

2.  **Modification:**
    * This stage is central to the Phase Vocoder's effectiveness. If one were to simply replay the STFT frames (the spectral snapshots) at a different rate for time stretching, it would cause phase mismatches, leading to "phasiness" or metallic sounds.
    * To avoid this, the algorithm estimates the **instantaneous frequency** for each frequency component. This is essentially how fast the phase of that component is changing from one analysis frame to the next.[<sup>[4]</sup>](#reference4)
    * When creating the output signal with a new timing (i.e., a different synthesis hop size), the phase for each frequency component in a new synthesis frame is not just taken from an analysis frame directly. Instead, it's calculated by taking the phase from the *previous synthesis frame* and advancing it based on the measured instantaneous frequency and the duration of the *new synthesis hop*. This careful propagation of phase ensures that the individual frequency components in the reconstructed signal connect smoothly, preserving the sound's naturalness as much as possible. [<sup>[5]</sup>](#reference5)
    * Mathematically, if $\phi_{s}(t - R_s)$ was the phase in the prior synthesis frame and $\omega_i$ is the instantaneous frequency from the analysis stage, the new phase $\phi_{s}(t)$ is updated by adding $\omega_i \cdot R_s$, where $R_s$ is the synthesis hop size. The magnitudes of the frequencies are typically taken from the original analysis frames or interpolated.

3.  **Synthesis:**
    * After the spectral information (magnitudes and newly calculated phases) for each frame is determined, an Inverse FFT (IFFT) is applied to each frame. This converts the modified frequency-domain data back into a time-domain signal segment.
    * Finally, these synthesized time-domain segments are combined, usually through an overlap-add procedure, to form the complete, modified output audio signal. [<sup>[3]</sup>](#reference3)

More advanced Phase Vocoder implementations often include techniques like "phase locking," which aim to maintain the phase relationships between different frequency bins that belong to the same harmonic component of a sound. This can further improve the perceived quality of the processed audio, particularly for musical tones. [<sup>[5]</sup>](#reference5)

  **Ans:**

  The phase vocoder (PV) is a widely spread technique for processing audio signals. It employs a short-time Fourier transform (STFT) analysis-modify-synthesis loop and is typically used for time-scaling of signals by means of using different time steps for STFT analysis and synthesis.
  the explanation of the **limitations of the Phase Vocoder** and the **artifacts** that can happen during **time stretching or pitch shifting**.

 **Limitations of the Phase Vocoder**

   **1. Phase Coherence Issues**: It analysis the sound in a thick, solid piece of something and tracking the phases of different frequencies, But it doesn’t keep the phases of related frequencies in sync, then the result can be a sound blurry we can  notice  with quick sounds like drum hits or speech consonants.

   **2. Poor Transient Handling**: The algorithm assumes the sound changes gradually, which works fine for smooth or steady sounds. But it suddenly struggles when the sharp sounds hit or a “t” sound in speech. These transients can lose their crispness and end up sounding mushy or flattened.

   **3. Time-Frequency Resolution Tradeoff**: The phase Vocoder is usead as the  **Short-Time Fourier Transform (STFT)**, which breaks the sound into time windows. The time windows are two types they are:

 * **Long windows** give better frequency details but can’t capture quick changes in time.
* **Short windows** give better timing precision because they analyze the sound over small time slices so they can accurately detect when drum hits. But the tradeoff is that these short slices don’t contain enough frequency information details to clearly distinguish between closely spaced pitches.

 ***4. Aliasing and Spectral Leakage**: When you break up a singal then you need to apply a window to each segment. If it doesn't done then carefully use the right window type or enough overlap, you can get unwanted effects like **aliasing** or **spectral leakage**

 ***5. Limited Quality in Pitch Shifting**: Pitch shift audio uses the phase vocoder, then it changes the pitch by altering freuency components of the sound, becuase it doesn't distinguish between pitch and formants.
 This can make someone’s voice sound cartoonish or robotic. More advanced techniques, like PSOLA or wavelet-based methods, are better at preserving those natural vocal characteristics.

 ***Artifacts You Might Hear During Time Stretching or Pitch Shifting**

 ***1. Phasiness or Smearing**: where the sound appears blurred. This happen because of the phase vocoder changes the tomhng and frequency of the audio and during the process, so it can lose the phase relationships between harmonics or partials.

 ***2. Transient Degradation**
You might notice that sharp, snappy sounds (like claps or drum hits) don’t hit as hard. That’s because the stretching spreads these fast events out in time, dulling their impact.

 ***3. Robotic or Metallic Sound**: In Vocals, the pitch is shifted one place another place then the result can be sound artficial (robot or metallic sound)
 This happens because the natural harmonic relationships in the sound are distorted.

 ***4. Chorus or Flanging Effects**: Sometimes, small differences in frequency across the frame can create interference patterns that sound like chorusb those swooshing effects that weren’t in the original.
